# colunas auxiliares

```python
rota['lat2'] = rota['lat'].shift(-1)
rota['long2'] = rota['long'].shift(-1)
rota['nome_local2'] = rota['nome_local'].shift(-1)
rota['trecho'] = rota['nome_local']+"----"+rota['nome_local2']
```

```python
display(rota)
```

```python
url = "https://route-and-directions.p.rapidapi.com/v1/routing"
```

```python
headers = {
	"x-rapidapi-key": f"{senha}",
	"x-rapidapi-host": "route-and-directions.p.rapidapi.com"
}
```

# cirando as chamadas

```python
list_chamadas = list()
for i in range(len(rota)-1):
    string = f"{rota['lat'][i]},{rota['long'][i]}|{rota['lat2'][i]},{rota['long2'][i]}"
    print(string)
    list_chamadas.append(string)
```

# ao solicitar, armazenamos as respostas numa lista para uso futuro

```python
list_responses = list()
```

```python
for waypoint in list_chamadas:
    querystring = {"waypoints":f"{waypoint}","mode":"drive"}
    response = requests.get(url, headers=headers, params=querystring)
    # print(response.json())
    list_responses.append(response.json())
```

### percorremos as respostas colhidas e armazenamos os dados em dois vetores com os pontos de lçatitudes e longitudes sobre as rodovias, brs, ruas

```python
latitudes = list()
longitudes = list()
for resposta in list_responses:
    for value in resposta['features'][0]['geometry']['coordinates'][0]:
        latitudes.append(value[1])
        longitudes.append(value[0])



# print(latitudes)
# print(longitudes)
```

## armazenando os dados de tempo e distancia definidos na api

```python
times = list()
for resposta in list_responses:
    times.append(resposta['features'][0]['properties']['time'])


distancias = list()
for resposta in list_responses:
    distancias.append(resposta['features'][0]['properties']['distance'])
```

```python
distancias.append(0)
```

```python
times.append(0)
```

```python
rota['tempo'] = times
rota['distancia'] = distancias
```

# definindo o arquivo coordinates

```python
coordinates = list()
for i in range(len(latitudes)):
    listauxiliar = list()
    listauxiliar.append(latitudes[i])
    listauxiliar.append(longitudes[i])
    coordinates.append(listauxiliar)
coordinates
```

# criando o rotograma

```python

# Create the map and add the line
m = folium.Map(location=[-18.100573, -40.522776], zoom_start=4)

folium.PolyLine(
    locations=coordinates,
    color="#FF0099",
    weight=5,
    tooltip="From Boston to San Francisco",
).add_to(m)


for i in range(len(rota)):
    folium.Marker(
        location=[f"{rota['lat'][i]}", f"{rota['long'][i]}"],
        tooltip="Click me!",
        popup=f"{rota['nome_local'][i]}",
        icon=folium.Icon(color="green"),
    ).add_to(m)


m
```
